```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#task1

df=pd.read_csv('cars (1) (1).csv')
```

```python
df.head(1)
```

```python
plt.figure(figsize=(30,10))
plt.plot(df['model'],df['hp'])
plt.xlabel('model')
plt.ylabel('hp')
plt.title('Model Name vs Horse Power')

plt.show()
```

```python
plt.bar(df['carb'],df['gear'],color='aqua')
plt.xlabel('Number of Carburaters')
plt.ylabel('Number of gear')
plt.title('Carbs vs gear')
plt.show()
```

```python
plt.hist(df['wt'],bins=30)
plt.xlabel('weight of the cars')
plt.ylabel('Count')
plt.title('Histogram for the weight value')
plt.show()
```

```python
#task4
cyl_value=df['cyl'].value_counts()
cyl_value
```

```python
plt.pie(cyl_value,autopct='%1.1f%%',explode=(0,0,0.05),shadow=True,labels=cyl_value,radius=1)
plt.title('Distribution of Cylinders')
plt.show()
```

```python
plt.pie(df['model'].value_counts(),labels=df['model'].value_counts().index,autopct='%1.1f%%')
plt.title('Distribution of Models')
plt.show()
```
